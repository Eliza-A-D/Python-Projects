# Do students in popular majors make more money?

`Median` = Median salary of full-time, year-round workers

```python
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter', title='Median vs. Share of Women', figsize=(5,2))
```

```python
recent_grads.plot(x='Median', y='Part_time', kind='scatter', title='Part Time Employees vs. Median', figsize=(5,2))
```

There isn't a substantial difference between the earnings of part-time employees and that of full-time employees.

The higher the number of full time employees the lower the median salary tends to be. This suggests that if a major has more full-time employees they have a lower median salary.

# Do Men or Women Earn More on Average?

```python
plt.plot(recent_grads['Median'],recent_grads['Women'], c='Purple')
plt.plot(recent_grads['Median'],recent_grads['Men'], c='Red')
```

Can group the data to answer this question.

```python
females = recent_grads[['Women', 'Median']]
print(females)
```

```python
total_women = females['Women'].sum()
```

```python
Total_salaries = []

for i in range(len(females)):
        a = females['Women'][i] * females['Median'][i]
        Total_salaries.append(a)


#females['Total_salaries'] = females['Women'] * females['Median']
```

```python
len(Total_salaries)
```

```python
len(females)
```

```python
females['Total_salaries'] = Total_salaries
```

```python
females
```

```python
average_female_salary = females['Total_salaries'].sum()/females['Women'].sum()
```

```python
round(average_female_salary,0)
```

```python
males = recent_grads[['Men', 'Median']]
total_men = males['Men'].sum()

Total_salaries = []

for i in range(len(males)):
        a = males['Men'][i] * males['Median'][i]
        Total_salaries.append(a)
        
males['Total_salaries'] = Total_salaries

average_male_salary = males['Total_salaries'].sum()/males['Men'].sum()
```

```python
round(average_male_salary,0)
```

The average male salary is `$40575`, which is higher than the average female salary of `$36300`
